```python
import sqlite3

conn = sqlite3.connect('cmpd_data.db')
c = conn.cursor()
```

```python
import pandas as pd

# load the data into a Pandas DataFrame
cmpd_data = pd.read_csv('CMPD_Incidents.csv', low_memory=False)
# write the data to a sqlite table
cmpd_data.to_sql('cmpd_data', conn, if_exists='append', index = False)
```
